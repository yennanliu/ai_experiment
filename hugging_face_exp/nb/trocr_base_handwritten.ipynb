```python
from transformers import TrOCRProcessor, VisionEncoderDecoderModel
from PIL import Image
from PIL import ImageEnhance # Import ImageEnhance directly
import requests


# Use the "printed" version of TrOCR
processor = TrOCRProcessor.from_pretrained('microsoft/trocr-base-printed')
model = VisionEncoderDecoderModel.from_pretrained('microsoft/trocr-base-printed')

# Use your image URL
url = "https://jeroen.github.io/images/testocr.png"
image = Image.open(requests.get(url, stream=True).raw).convert("RGB")

# Resize to improve OCR (TrOCR expects 384x384 by default)
image = image.resize((384, 384))

# Improve contrast
enhancer = ImageEnhance.Contrast(image) # Use ImageEnhance directly
image = enhancer.enhance(2.0)  # increase contras

# Prepare inputs
pixel_values = processor(images=image, return_tensors="pt").pixel_values

# Run inference
generated_ids = model.generate(pixel_values)
generated_text = processor.batch_decode(generated_ids, skip_special_tokens=True)[0]

print("Recognized Text:", generated_text)
```

```python
from transformers import TrOCRProcessor, VisionEncoderDecoderModel
from PIL import Image, ImageEnhance
import requests

# Step 1: Load image
url = "https://jeroen.github.io/images/testocr.png"
image = Image.open(requests.get(url, stream=True).raw).convert("RGB")

# Step 2: Resize to match TrOCR expectations
image = image.resize((384, 384))

# Step 3: Enhance contrast
enhancer = ImageEnhance.Contrast(image)
image = enhancer.enhance(2.0)  # You can try values like 1.5, 2.0, 2.5

# Optional: Uncomment to convert to grayscale
# image = image.convert("L")

# Step 4: Load processor and model (printed TrOCR)
processor = TrOCRProcessor.from_pretrained("microsoft/trocr-base-printed")
model = VisionEncoderDecoderModel.from_pretrained("microsoft/trocr-base-printed")

# Step 5: OCR inference
pixel_values = processor(images=image, return_tensors="pt").pixel_values
generated_ids = model.generate(pixel_values)
generated_text = processor.batch_decode(generated_ids, skip_special_tokens=True)[0]

print("Recognized Text:", generated_text)
```

```python
from transformers import Blip2Processor, Blip2ForConditionalGeneration
import torch
from PIL import Image
import requests

# Load BLIP-2 model and processor
processor = Blip2Processor.from_pretrained("Salesforce/blip2-flan-t5-xl")
model = Blip2ForConditionalGeneration.from_pretrained("Salesforce/blip2-flan-t5-xl")

# Load image
url = "https://jeroen.github.io/images/testocr.png"
image = Image.open(requests.get(url, stream=True).raw).convert("RGB")

# Prepare inputs
inputs = processor(image, "Please extract all the text in the image.", return_tensors="pt")

# Generate text
generated_ids = model.generate(**inputs)
text = processor.decode(generated_ids[0], skip_special_tokens=True)

print("Extracted Text:", text)
```
