```python
import requests
from bs4 import BeautifulSoup
from datetime import datetime, timedelta
```

```python
BASE_URL="https://www.twblogs.net"
url_list = []
html_content = []
title_list = []
```

```python
# URL of the website
url = "https://www.twblogs.net/c/5b7a8baa2b7177392c96436d/"

# Send a GET request to the URL
response = requests.get(url)

# Check if the request was successful (status code 200)
if response.status_code == 200:
    # Parse the HTML content of the response
    soup = BeautifulSoup(response.content, "html.parser")
    #print("soup = " + str(soup))

    elements = soup.find_all("a", class_="", attrs={"data-post-id": True})
    
    for element in elements:
        print ("Title = " + str(element.text.strip()))
        print ("Href = " +  str(element["href"]))
        print("data-post-id = " + str(element["data-post-id"]))
        print ("")
        url_list.append(BASE_URL + element["href"])
        title_list.append(element.text.strip())

    # Extract and print the text content of each element
#     for element in elements:
#         print("""
#         Title = {title}
#         """.format(title = element.text.strip())
#         #print(">>> element = " + str(element["href"]))

else:
    print("Failed to retrieve data from the website")
```

```python
idx = 0
for _url in url_list:
    
    response = requests.get(_url)
    if response.status_code == 200:
        print ("------------------> Start " + str(idx))
        print ("_url = " + str(_url))
               
        # Parse the HTML content of the response
        soup = BeautifulSoup(response.content, "html.parser")
        #print("soup = " + str(soup))
        #elements = soup.find_all("a", class_="", attrs={"data-post-id": True})
        #div_tag = soup.find('div', class_='blogpost-body blogpost-body-html', id='cnblogs_post_body')

        # Extract all content inside <p> tags within the <div> tag
        paragraphs = soup.find_all('p')
        
        if paragraphs != None:

            # Extract text from each <p> tag
            content = '\n'.join([p.get_text() for p in paragraphs])

            print (content)
            html_content.append(content)

        # Extract the content inside the <div> tag
#         if div_tag != None:
#             content = div_tag.get_text()
#             print(content)
        idx += 1
        print ("------------------> End")

    else:
        print("Failed to retrieve data from the website")
```

```python
# Define the base header template
BASE_HEADER = """
+++
title = '{title}'
date = {timestamp}
draft = false
+++
"""

more="<!--more-->"
```

```python
today = datetime.now()
yesterday = today - timedelta(days=1)
current_time = yesterday.strftime("%Y-%m-%dT%H:%M:%S%z")
    
for k, _content in enumerate(html_content):
    # 2024-02-10T09:22:22+08:00
    header = BASE_HEADER.format(title=title_list[k], timestamp=current_time)
    result = header + more + _content
    print (result)
    file_name = title_list[k] + ".md"
    with open("../../hugo-blog/quickstart/content/posts/" + file_name, "w") as file:
        file.write(result)
        print("{} file has been created.".format(file_name))
        print ("-------------------")
```
