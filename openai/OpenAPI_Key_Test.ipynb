# Get Prompt Method

```python
def get_prompt_resp(prompt, role_description, temperature=0.2):
  try:
      print("----> prompt = " + str(prompt))
      print("----> role_description = " + str(role_description))
      response = client.chat.completions.create(
          model="gpt-4",
          messages=[
              {"role": "system", "content": role_description},
              {"role": "user", "content": prompt},
          ],
          temperature=temperature,  # deterministic, good for coding tasks
      )

      print("‚úÖ API key is working. Model responded with:")
      print(response.choices[0].message.content)

  except Exception as e:
      print("‚ùå API key test failed.")
      print("Error:", str(e))
```
