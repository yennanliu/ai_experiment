```python
prompt = """
請提供ㄧ個有6個章節的科幻小說,
每個章節結尾請用 ++++ 分隔
標題：《星際之旅》
角色: John, Amy, 元素: 蟲洞, 重力, 相對座標, 太空船
"""

# Call the ChatGPT API using the new ChatCompletion interface
response = openai.Completion.create(
    engine="gpt-3.5-turbo-instruct",  # Use the appropriate engine for ChatGPT
    prompt=prompt,
    max_tokens=3000  # You can adjust this parameter as needed
)

decoded_response = response['choices'][0]['text'].encode('utf-8').decode('utf-8')
print(decoded_response)
```

```python
decoded_response = response['choices'][0]['text'].encode('utf-8').decode('utf-8')
print(decoded_response)
```

```python
chapters = decoded_response.split("+++")
```

```python
for k in range(len(chapters)):
    prompt = """
    請根據以下大綱當作故事骨架, 提供至少1000字的內容, 
    用寫實筆法描寫 並加入創新元素
    大綱:
    {content}
    """
    print ("k=" + str(k))
    #print (chapters[k])
    print(chapters[k])
    content=chapters[k].replace("+","")
    prompt=prompt.format(content=content)
    #print (prompt)
    
    # Call the ChatGPT API using the new ChatCompletion interface
    _response = openai.Completion.create(
        engine="gpt-3.5-turbo-instruct",  # Use the appropriate engine for ChatGPT
        prompt=prompt,
        max_tokens=3000  # You can adjust this parameter as needed
    )

    _decoded_response = _response['choices'][0]['text'].encode('utf-8').decode('utf-8')
    print("-------->")
    print(_decoded_response)
    print()
```

```python
# https://www.mgyqw.com/mgyqw/21603.html


prompt = """
請提供ㄧ個有7個章節的愛情小說,
請提供標題, 大綱, 章節
每個章節結尾請用 ++++ 分隔
並根據以下情境發揮:


結婚兩年多，她對于自家老公還是不太了解，
只知道他是建設公司總經理，以及……桃花亂開的萬人迷，
誰叫他們是相親後閃電結婚，沒啥感情基礎，
而且彼此生活習慣不同，又各自忙碌，更沒什麼機會談心，
看樣子，他們的婚姻生活會這樣平淡無趣下去……
等等，這男人怎麼會出現在咖啡店，
不只難得親熱地叫她老婆，替她擺脫客戶的殷勤邀約，
還說她展現出的律師專業神采，再次迷住了他，
更詭異的是，等她察覺時，她已習慣和他分享生活大小事，
他不過是出差三、四天，一向獨立的她就寂寞得難以入眠，
知道他加班超過三天，她便忙著炖雞湯替他補元氣，
更別提得知他出車禍，理智負責的她竟丟下工作沖去醫院，
她想，她是在不知不覺中喜歡上他了，
只是那些來探病的美女，令她很有收回這份感情的沖動……
"""

# Call the ChatGPT API using the new ChatCompletion interface
response = openai.Completion.create(
    engine="gpt-3.5-turbo-instruct",  # Use the appropriate engine for ChatGPT
    prompt=prompt,
    max_tokens=3000  # You can adjust this parameter as needed
)

decoded_response = response['choices'][0]['text'].encode('utf-8').decode('utf-8')
print(decoded_response)
```
